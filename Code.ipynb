```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import seaborn as sns
from matplotlib import pyplot
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab
import matplotlib 
%matplotlib inline
color = sns.color_palette()
import matplotlib.ticker as mtick
```

```python
df = pd.read_excel('Elenco_Ticket_19_07_2022 15_45_39.xlsx')
df1 = df.drop('Recapito', axis=1)
df1 = df1.drop('Ticket master', axis = 1)
df1 = df1.drop('Richiedente', axis = 1)
df1 = df1.drop('In carico a', axis= 1)
df1 = df1.drop('Aperto da', axis = 1)
df1 = df1.drop('Utente Chiusura', axis = 1)
df1 = df1.drop('Data Chiusura', axis = 1)
df1 = df1.drop('Data di creazione', axis = 1)
df1 = df1.drop('Note', axis = 1)
df1 = df1.drop("Societa'", axis = 1)
df1 = df1.drop('Numero', axis = 1)
df1
```

```python
df1.info()
df2 = df1.dropna(axis=0)
df2
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df2['Sede']= label_encoder.fit_transform(df2['Sede'])
df2['U.O.C. di Riferimento']= label_encoder.fit_transform(df2['U.O.C. di Riferimento'])
df2['Classificazione']= label_encoder.fit_transform(df2['Classificazione'])
df2['Attività']= label_encoder.fit_transform(df2['Attività'])
df2['Area di riferimento']= label_encoder.fit_transform(df2['Area di riferimento'])
df2['Stato']= label_encoder.fit_transform(df2['Stato'])
df2
```

```python
fig = plt.figure(figsize=(25, 30))
i = 0
for column in df2:
    sub = fig.add_subplot(7, 7, i+1)
    sub.set_xlabel(column)
    df2[column].plot(kind = 'hist')
    i = i+1
```

```python
sns.set(style="white")
corr = df2.corr()
plt.figure(figsize = (15,15))
sns.heatmap(corr, cmap='coolwarm', annot=True)
```

```python
y = df2['Attività']
x = df2.drop('Attività', axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler 
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, fbeta_score, roc_auc_score
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(x_train,y_train)
y_predLOG = logreg.predict(x_test)
```

```python
acc = accuracy_score(y_test, y_predLOG)
print(acc)
```

```python
from sklearn.ensemble import RandomForestClassifier
score_array = []
for each in range(1,100):
    rf_loop = RandomForestClassifier(n_estimators = each, random_state = 1) 
    rf_loop.fit(x_train,y_train)
    score_array.append(rf_loop.score(x_test,y_test))
 
fig = plt.figure(figsize=(15, 7))
plt.plot(range(1,100),score_array, color = '#ec838a')
plt.ylabel('Range\n',horizontalalignment="center",
fontstyle = "normal", fontsize = "large", 
fontfamily = "sans-serif")
plt.xlabel('Score\n',horizontalalignment="center",
fontstyle = "normal", fontsize = "large", 
fontfamily = "sans-serif")
plt.title('Optimal Number of Trees for Random Forest Model \n',horizontalalignment="center", fontstyle = "normal", fontsize = "22", fontfamily = "sans-serif")
plt.grid(label=True)
plt.xticks((np.arange(0, 100, 5.0) ),rotation=0, horizontalalignment="center")
plt.yticks(rotation=0, horizontalalignment="right")
plt.show()
```

```python
clf=RandomForestClassifier(n_estimators=63)
clf.fit(x_train,y_train)
y_predRF = clf.predict(x_test)
```

```python
acc1 = accuracy_score(y_test, y_predRF)
print(acc1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
score_array = []
for each in range(1,25):
    knn_loop = KNeighborsClassifier(n_neighbors = each) 
    knn_loop.fit(x_train,y_train)
    score_array.append(knn_loop.score(x_test,y_test))
fig = plt.figure(figsize=(15, 7))
plt.plot(range(1,25),score_array, color = '#ec838a')
plt.ylabel('Range\n',horizontalalignment="center",
fontstyle = "normal", fontsize = "large", 
fontfamily = "sans-serif")
plt.xlabel('Score\n',horizontalalignment="center",
fontstyle = "normal", fontsize = "large", 
fontfamily = "sans-serif")
plt.title('Optimal Number of K Neighbors \n',
horizontalalignment="center", fontstyle = "normal", fontsize = "22", fontfamily = "sans-serif")
plt.xticks((np.arange(0, 26, 2.0)), rotation=0, horizontalalignment="center")
plt.yticks(rotation=0, horizontalalignment="right")
plt.grid(label=True)
plt.show()
```

```python
knn = KNeighborsClassifier(n_neighbors=14)
knn.fit(x_train, y_train)
y_predKNN = knn.predict(x_test)
```

```python
acc2 = accuracy_score(y_test, y_predKNN)
print(acc2)
```

```python
from sklearn import svm
clf1 = svm.SVC(kernel='linear') 
clf1.fit(x_train, y_train)
y_predSVM = clf1.predict(x_test)
```

```python
acc3 = accuracy_score(y_test, y_predSVM)
acc3
```
